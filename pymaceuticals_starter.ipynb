```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
merge_df=pd.merge(mouse_metadata,study_results,on="Mouse ID",how="left")
merge_df.head()
```

```python
# Checking the number of mice in the DataFrame.
len(merge_df.value_counts(("Mouse ID")))
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
drop_dup_mouse_id = merge_df.loc[merge_df.duplicated(subset=['Mouse ID', 'Timepoint',]),'Mouse ID'].unique()
drop_dup_mouse_id 
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_mouse_df = merge_df[merge_df['Mouse ID'].isin(drop_dup_mouse_id)==False]
clean_mouse_df.sort_values("Timepoint", ascending= True)
clean_mouse_df.head()
```

```python
# Checking the number of mice in the clean DataFrame.
len(clean_mouse_df.value_counts(("Mouse ID")))
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
```

```python
corr=round(st.pearsonr(avg_capm_vol['Weight (g)'],avg_capm_vol['Tumor Volume (mm3)'])[0],2)
print(f"The correlation between mouse weight and average tumor volume is {corr}")
x_values = avg_capm_vol['Weight (g)']
y_values = avg_capm_vol['Tumor Volume (mm3)']

(slope, intercept, rvalue, pvalue, stderr) = st.linregress(x_values, y_values)
regress_values = x_values * slope + intercept

print(f"slope:{round(slope,2)}")
print(f"intercept:{round(intercept,2)}")
print(f"rvalue (Correlation coefficient):{round(rvalue,2)}")
print(f"pandas (Correlation coefficient):{corr}")
print(f"stderr:{round(stderr,2)}")

line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

print(line_eq)

fig1, ax1 = plt.subplots(figsize=(15, 10))
plt.scatter(x_values,y_values,s=175, color="coral",marker="*")
plt.plot(x_values,regress_values,"r-",color="lightblue")
plt.xlabel('Weight(g)',fontsize =14)
plt.ylabel('Average Tumore Volume (mm3)',fontsize =14)
ax1.annotate(line_eq, xy=(20, 40), xycoords='data',xytext=(0.8, 0.95), textcoords='axes fraction',horizontalalignment='right', verticalalignment='top',color="blue",fontsize=20)
print(f"Coefficient of determination: {round(rvalue**2,2)}")
plt.title('Mouse weight versus average tumor volume for the Capomulin regimen',fontsize=25)
plt.grid()
plt.show()
```
